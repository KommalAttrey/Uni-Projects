```python
import ast
# Import the ast module for safe evaluation of string representations
def extract_text_original(snippet):
    snippet_dict = ast.literal_eval(snippet)

    # Extracting the 'textOriginal' field
    if 'topLevelComment' in snippet_dict:
        comment_snippet = snippet_dict['topLevelComment']['snippet']
        return comment_snippet.get('textOriginal', '')
    else:
        return ''

# Create a copy of the DataFrame
original_df = combined_df.copy()

# Apply the function to original_df and adding the a column into the dataframe
original_df['textOriginal'] = original_df['snippet'].apply(extract_text_original)
print(original_df.textOriginal.head(50))
```

#Explanation and Justification
Order of preprocessing

Here we are trying to refine the text data by removing less informative and irrelevant tokens at each stage. This approach helps to reduce the dimensionality of the data, and focus on the most meaningful tokens for analysis.

*  Remove context-independent stopwords:
Stopwords are common words ( not carry significant meaning in the context).Removing them early in the preprocessing pipeline helps to reduce the size of text data which will improve efficiency as subsequent preprocessing steps can operate on a smaller set of tokens.

* Remove short tokens:
Short tokens(length less than 3)  are often not informative and to focus on more meaningful tokens and to remove noise in the analysis this step has been performed second.Also it has been done after removing stopwords because some stopwords might also be short tokens, and removing them first avoids unnecessary processing.

*   Remove rare tokens:
Rare tokens are words that appear infrequently and may not have a significant impact overall. We are performing it after removing short tokens because  here rare tokens(based on their frequency) removing short tokens first provides a more accurate representation of token frequencies.Keeping short tokens during this step could potentially skew the frequency calculations and lead to incorrect identification of rare tokens.so small tokens removed before this.

*  Remove context-dependent stopwords:
Words which are common within a specific context and might not be relevant for the analysis.(do not contribute to the generally).By removing rare tokens first, the frequency calculations for context-dependent stopwords will be more accurate and meaningful.

*   Stemming - Process of reducing words to their base or root form.
Stemming helps to consolidate different forms of a word into a single representation, reducing the dimensionality of the text data.
We are performing it at last because as stemming should be applied on refined set of tokens after removing and processing steps. Also to focus on more elevant and informative tokens.

*   Include bigrams - After stemming bigrams has been generated by the final tokens we have obtained.
*  Combine unigrams and bigrams - both of these has been combined and stored as a new column in the dataframe.

```python
import re

# Define the regular expression pattern for tokenization
token_pattern = r"[a-zA-Z]+"

def tokenize(text):
    return re.findall(token_pattern, text)

# Apply tokenization to the 'textOriginal' column
# Adding tokens as a column in the dataframe
filtered_df = filtered_df.copy()
filtered_df['tokens'] = filtered_df['textOriginal'].apply(tokenize)
print(filtered_df['tokens'].head(10))
```

#2
*   Context-independent stopwords are the words which occurs frequently across all documents but convey very little information abut the content of the document.
*  Here we are storing the list of context-independent stopwords from the text file into a list named stopwords.
*  We are defining a function which will take list of tokens as input then we are iterating over each token to check if it is present in the stopword list or not.
*  Applying this function to the 'tokens' column.





```python
#2

with open("/content/drive/Shareddrives/FIT5196_S1_2024/A1/stopwords_en.txt", "r") as file: # Load the list of stopwords from given file
    stopwords = file.read().splitlines()

# Defining a function to remove stopwords
def remove_stopwords(tokens):
    if isinstance(tokens, list):
        return [token for token in tokens if token.lower() not in stopwords]
    else:
        return tokens


# Apply function on 'stemmed_tokens' column
filtered_df['filter_context_independent_tokens'] = filtered_df['tokens'].apply(remove_stopwords)
print(filtered_df[['tokens', 'filter_context_independent_tokens']].head(10))
```

#3

*  As we thought very short words do not contribute much to the overall meaning of the text and can be remove here early in the preprocessing pipeline.
*   So to focus on more important and informative content tokens shorter than length 3 are removed here.
*   Here applying a lambda function to the 'tokens' column of the filtered dataframe. This lambda function iterates over each list of tokens and filters out tokens with a length less than 3





```python
#3
#Remove tokens with a length less than 3
filtered_df.loc[:, 'filter_context_independent_tokens'] = filtered_df['filter_context_independent_tokens'].apply(lambda x: [token for token in x if len(token) >= 3])
print(filtered_df['filter_context_independent_tokens'].head(20))
```

#4


*   Rare Token Removal:
In this step, we are removing rare tokens that appear infrequently across the channel IDs.
*   Here we are counting the occurrences of each token across all channel IDs using the Counter class from the collections module. The token_occurrences Counter object keeps track of the count of each token.



*   We calculate the threshold for rare tokens based on channels_ids that have at least 15 english comments and threshold being defined.
*   We iterate over each token and its count in the token_occurrences Counter. If the count is less than the threshold, we append the token to the rare_tokens list. then converting that list to set to remove dupliocates and finally,applying a lambda function to the 'filter_context_independent_tokens' column of the filtered_df DataFrame.

```python
#4
from collections import Counter

token_occurrences = Counter()
for tokens in filtered_df['filter_context_independent_tokens']: # Count the occurrences of each token across all channel IDs
    token_occurrences.update(set(tokens))

total_unique_channel_ids = len(filtered_df['channel_id'].unique()) # Calculating threshold for rare tokens
threshold = int(0.01 * total_unique_channel_ids)

rare_tokens = []
for token, count in token_occurrences.items():  # Filter tokens that appear in less than 1% of unique channel IDs
    if count < threshold:
        rare_tokens.append(token)
rare_tokens_set = set(rare_tokens)

#Removing rare tokens from the tokens list
filtered_df['rare_tokens_removed'] = filtered_df['filter_context_independent_tokens'].apply(
    lambda tokens: [token for token in tokens if token not in rare_tokens_set]
)
filtered_df
```

#5


*    Context-dependent stopwords are words that might be meaningful in some contexts but not in others.These words may not be informative for distinguishing between different channels and are therefore removed to improve the quality of the vocabulary.
*  removing context-dependent stopwords that appear frequently across the channel IDs.(Similar to the rare token removal step)
    

*   counting the occurrences of each token across all channel IDs then calculating the threshold
*   Iterating over each token and its count in the token_occurrences Counter. If the count is greater than the threshold, we append the token to the context_dependent_stopwords list then defining a function  which will remove context-dependent stopwords. and applying it to previous column.

```python
#5
from collections import Counter
token_occurrences = Counter()
for tokens in filtered_df['rare_tokens_removed']: # Count the occurrences of each token across all channel IDs
    token_occurrences.update(set(tokens))

total_unique_channel_ids = len(filtered_df['channel_id'].unique())# Calculate the threshold for context-dependent stopwords
threshold = int(0.99 * total_unique_channel_ids)

context_dependent_stopwords = []
for token, count in token_occurrences.items():  # Filter tokens that appear in more than 99% of unique channel IDs
    if count > threshold:
        context_dependent_stopwords.append(token)
        print(f"Token: '{token}', Count: {count}, Threshold: {threshold}")

print("\nContext-dependent stopwords:")
print(context_dependent_stopwords)


def remove_context_dependent_stopwords(tokens):
    if isinstance(tokens, list):
        return [token for token in tokens if token not in context_dependent_stopwords] #Removing context-dependent stopwords
    else:
        return tokens

# Applying the function to rare_tokens_removed column and generating new column
filtered_df['context_dependent_tokens_removed'] = filtered_df['rare_tokens_removed'].apply(remove_context_dependent_stopwords)
```

```python
#Step-6
import nltk
from nltk.stem import PorterStemmer
porter_stemmer = PorterStemmer() #Initialize the Porter stemmer

# Defining a function to stem tokens
def stem_tokens(tokens):
    return [porter_stemmer.stem(token) for token in tokens]

filtered_df.loc[:, 'final_tokens'] = filtered_df['context_dependent_tokens_removed'].apply(stem_tokens)
print(filtered_df[['context_dependent_tokens_removed', 'final_tokens']].head(10))
```

#7
Bigrams capture meaningful associations between adjacent words and can provide additional context that single words might miss.

Converting Tokens to Flattened List

  * We converted the list of lists of tokens (tokens_list) into a single flattened list (tokens_flat).
  
Creating BigramCollocationFinder:
  *   We create a BigramCollocationFinder named finder from the flattened list of tokens. This finder will be used to identify bigrams.
Scoring Bigrams using PMI:

PMI
 *   To find the association strength between words in the bigrams We score the bigrams using Pointwise Mutual Information (PMI) through the score_ngrams method of the BigramCollocationFinder.
 *   Then sorting the bigrams based on their PMI scores in descending order.


Selecting Bigrams:

*   Here we are checking if both words occur together as adjacent words within the same comment. For that we are Iterating over the sorted bigrams and filter them to ensure they can be collocated within the same comment.


```python
import nltk
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures

tokens_list = filtered_df['final_tokens'].tolist()
tokens_flat = [token for sublist in tokens_list for token in sublist] # Flatten the list of lists into a single list of tokens


# Create a BigramCollocationFinder from the flattened list of tokens
finder = BigramCollocationFinder.from_words(tokens_flat)

# Score the bigrams using PMI (Pointwise Mutual Information)
bigram_scores = finder.score_ngrams(BigramAssocMeasures.pmi)
sorted_bigrams = sorted(bigram_scores, key=lambda x: x[1], reverse=True) #sorting
# Filter the selected bigrams to ensure they can be collocated within the same comment

selected_bigrams = [] #Initialize a list to store the selected bigrams along with their PMI scores
for bigram, pmi in sorted_bigrams:
    # Check if both words in the bigram occur together as adjacent words within the same comment
    for tokens in tokens_list:
        if len(tokens) >= 2:
            # Check if the bigram occurs as adjacent words within the same comment
            if bigram[0] in tokens and bigram[1] in tokens:
                index_1 = tokens.index(bigram[0])
                index_2 = tokens.index(bigram[1])
                if abs(index_1 - index_2) == 1:
                    selected_bigrams.append((bigram, pmi)) #Appending the bigram along with its PMI score.
                    break  # Move to the next bigram after finding a match
    # Check if 200 bigrams are selected
    if len(selected_bigrams) == 200:
        break

for bigram, pmi in selected_bigrams[:20]:
    print(bigram, pmi)
```

#8
Calculate the vocabulary containing both unigrams and bigrams.

*   The provided code iterates through each row in the DataFrame filtered_df, column 'final_tokens'.
 For each row, it forms combined tokens and bigrams, where adjacent tokens are combined with an underscore to create a bigram.
*   It checks if these combined tokens exist in the list of selected bigrams. If a bigram is found, it adds both the individual tokens and the bigram to the list.If not, it adds only the first token. and iterates through next tokens and so on.And generating a new column named 'combined_tokens_and_bigrams' in the DataFrame.


```python
combined_tokens_and_bigrams = []      # Initialize a list to store combined tokens and bigrams for each row

for tokens in filtered_df['final_tokens']: # Iterate through each row in the DataFrame
    combined_tokens_and_bigrams_row = []  # Initialize a list to store combined tokens and bigrams for this row
    i = 0
    while i < len(tokens):
        # Check if there are at least two tokens remaining to form a bigram
        if i < len(tokens) - 1:
            combined_token = tokens[i] + '_' + tokens[i + 1]     # Combine two adjacent tokens with an underscore to form a bigram
            for bigram, _ in selected_bigrams:                   # Check if the combined token exists in the list of selected bigrams
                if combined_token == '_'.join(bigram):
                    # If the combined token is in the list of selected bigrams, add both the individual tokens and the bigram to the list
                    combined_tokens_and_bigrams_row.append(tokens[i])
                    combined_tokens_and_bigrams_row.append(tokens[i + 1])
                    combined_tokens_and_bigrams_row.append(combined_token)
                    i += 2  # Move to the next pair of tokens
                    break
            else:
                # If the combined token is not a bigram, add both first tokens to the list
                combined_tokens_and_bigrams_row.append(tokens[i])
                i += 1
        else:
            # If there is only one token remaining, add it to the list
            combined_tokens_and_bigrams_row.append(tokens[i])
            i += 1  # Move to the next token

    # Append the combined tokens and bigrams list for this row to the main list
    combined_tokens_and_bigrams.append(combined_tokens_and_bigrams_row)



filtered_df['combined_tokens_and_bigrams'] = combined_tokens_and_bigrams # Adding new column in the DataFrame
#print(filtered_df[['final_tokens', 'combined_tokens_and_bigrams']])
```

#9
Combine unigrams and bigrams into vocabulary


*   A set named 'vocabulary' is initialized to encompass both unigrams and bigrams.

*   Unigrams, individual tokens, are iteratively added to the vocabulary from the 'final_tokens'column.
*   Bigrams added into the vocabulary after conversion to strings. Each bigram's words are joined by underscores to facilitate distinction.


```python
# Combine unigrams and bigrams
unigrams = filtered_df['final_tokens'].explode().tolist()  # Flatten the list of lists of unigrams
bigrams = [bigram[0] + '_' + bigram[1] for bigram, _ in selected_bigrams]  # Extract bigrams from the selected_bigrams list
vocabulary = set(unigrams + bigrams)

# Print the vocabulary and its count
print("Vocabulary:",vocabulary)
print("Vocabulary Count:", len(vocabulary))

vocabulary = [str(token) for token in vocabulary]
vocabulary_sorted = sorted(vocabulary) # Sort the vocabulary alphabetically


# Write the sorted vocabulary to a text file
with open('65_vocab.txt', 'w') as file:
    for index, token in enumerate(vocabulary_sorted, start=0):
        file.write(f"{token}:{index}\n")
```

#2



*   Importing the FreqDist class from the nltk.probability module to create a frequency distribution of tokens.
* The list combined_tokens_and_bigrams  is flattened into a single list.


*  Calculating frequency distribution of tokens in the flattened list using FreqDist()


```python
import nltk
from nltk.probability import FreqDist

# Flatten the list of lists of tokens and bigrams
tokens_flat = [token for sublist in filtered_df['combined_tokens_and_bigrams'] for token in sublist]

# Use FreqDist() to count the frequency of each token
freq_dist = FreqDist(tokens_flat)

for token, frequency in freq_dist.items():
  print(f"{token}: {frequency}")
```

#3
*   The code creates a dictionary named token_to_index to map tokens to their corresponding indices in the vocabulary.
*  It uses a dictionary comprehension to iterate over the vocabulary_sorted list, where each token is paired with its index.
*   Iterate through each token list in the 'combined_tokens_and_bigrams' column of the filtered_df DataFrame.Initialize an empty list and mapped_tokens to store the mapped tokens and their indices for the current token list.


```python
# Create a dictionary to map tokens to their indices in the vocabulary
token_to_index = {token: index for index, token in enumerate(vocabulary_sorted)}

# Iterate through the tokens in the combined_tokens_and_bigrams column
for token_list in filtered_df['combined_tokens_and_bigrams']:
    mapped_tokens = []
    for token in token_list:
        # Check if the token exists in the vocabulary
        if token in token_to_index:
            # Map the token to its corresponding index
            index = token_to_index[token]
            mapped_tokens.append((token, index))

    #print(mapped_tokens)
```

#4
* Iterating through the comments (rows) of the DataFrame filtered_df, simultaneously iterating through the corresponding tokenized texts stored in combined_tokens_and_bigrams column of dataframe.

*   For each comment, it writes the associated channel ID to the file, followed by a comma & It initializes a list named token_freq_pairs to store token index and frequency pairs.
*   For each token in the comment It checks if the token exists in the token_to_index mapping. if yes it retrieves its index and frequency.
and appends the token index and frequency pair to the token_freq_pairs list.

*   Then writing the token index and frequency pairs to the file, separated by commas.






```python
with open("65_countvec.txt", "w") as f:
    # Iterate through the comments (rows)
    for channel_id, tokens in zip(filtered_df['channel_id'], filtered_df['combined_tokens_and_bigrams']):
        # Initialize a list to store token index and frequency pairs
        token_freq_pairs = []

        # Check if the token list is not empty
        if tokens:
            for token in tokens:
                if token in token_to_index:
                    index = token_to_index[token]
                    frequency = freq_dist[token]
                    token_freq_pairs.append(f"{index}:{frequency}")

        # Write the channel ID and token index and frequency pairs to the file
        if token_freq_pairs:
            f.write(f"{channel_id},{','.join(token_freq_pairs)}\n")
        else:
            f.write(f"{channel_id}\n")
```
